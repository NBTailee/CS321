```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd

train_file_path = r"C:\Users\leduc\OneDrive\Desktop\NLP\CS321\SA\VLSP2018-SA-train-dev-test\train.csv"
test_file_path = r"C:\Users\leduc\OneDrive\Desktop\NLP\CS321\SA\VLSP2018-SA-train-dev-test\test.csv"
df_train = pd.read_csv(train_file_path, encoding="utf-8")
df_test = pd.read_csv(test_file_path, encoding="utf-8")


df_train = df_train[:800]

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2), max_features=5000)),
    ('clf', MultiOutputClassifier(LogisticRegression(max_iter=1000, solver='liblinear')))
])


X_train = df_train["Review"]
X_test = df_test["Review"]


y_train = df_train.drop(columns=["Review"])
y_test = df_test.drop(columns=["Review"])


pipeline.fit(X_train, y_train)


y_pred = pipeline.predict(X_test)


y_test = np.array(y_test)
y_pred = np.array(y_pred)

```
